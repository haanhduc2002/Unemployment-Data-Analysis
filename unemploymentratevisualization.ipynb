Unpivor tables for visualization


```python
#Create a list of years 
years = list(df.columns)
del years[0:2]
#Unpivot the current dataset 
df_unpivot  = pd.melt(df, id_vars=["Country Name", "Country Code"], value_vars= years, var_name= "Year", value_name= "Unemployment rate")
df_unpivot
```

After viewing the dataset, non-country name are found. I extracted the country code and non-country code using pycountry then stores them in seperated list

```python
region_codes = []
country_codes= []


def generate_Codes(region_codes:list, country_codes:list):
    #Create a list of country codes
    codes = df["Country Code"].tolist()
    countries = []

    for country in list(pycountry.countries):
        countries.append(country.alpha_3)  # Creates a list full of official country names

    for code in codes:
        if code not in countries:
            region_codes.append(code)
        else:
            country_codes.append(code)

generate_Codes(region_codes, country_codes)
```

After code extraction we can see that, the dataset have provided not only unemployment rate by countries but also by categories, for instance divided by regions, divided by economic status and divided by geography.

Data visualization using plotly.graph_object

Visualize all categaries' unemployment from 1991 to 2021 (non-country code)

```python
#line chart
def generate_Line(nations_codes):
    fig  = go.Figure()

    for nation in nations_codes:
        fig.add_trace(
            go.Scatter(
                x= df_unpivot[df_unpivot["Country Code"]== nation]["Year"],
                y= df_unpivot[df_unpivot["Country Code"]== nation]["Unemployment rate"], 
                name = df[df["Country Code"]== nation]["Country Name"].values[0], 
                ))

    fig.update_layout(
        title_text = "Unployment rate from 1991 to 2021 by categories",
        xaxis_title='Year',
        yaxis_title='Unemployment rate(%)', 
        legend_itemclick = "toggleothers", 
        legend_itemdoubleclick = "toggle",
        plot_bgcolor="#FFF",  # Sets background color to white
        xaxis=dict(
            title="time",
            linecolor="#BCCCDC",  # Sets color of X-axis line
            showgrid=False,  # Removes X-axis grid lines
            showspikes=True, # Show spike line for X-axis
            # Format spike
            spikethickness=2,
            spikedash="dot",
            spikecolor="#999999",
            spikemode="across",
        ),
        yaxis=dict(
            title="price",  
            linecolor="#BCCCDC",  # Sets color of Y-axis line
            showgrid=False,  # Removes Y-axis grid lines    
        ))

    fig.update_layout(xaxis= dict(rangeslider = dict(visible= True)))

    return fig


fig = generate_Line(region_codes)
fig.show()
```

In general, world unemployment rates appear to be decreasing before recession and pandemic happens and increase right after these events happened (the recession in 2008 and pandemic in late 2019). High income countries are the ones who suffer the most during the pandemics and the 2008 recession but they manage to adapt and lower their unemployment rate much faster than middle to low income countries.

Visualize countries' unemployment rate in specific year

```python
def generate_Choropleth(year):
    #convert to string
    year= str(year)
    
    #validate year
    if year not in df.columns:
        return go.Figure()
    else:
        fig = go.Figure(data = go.Choropleth(
            locations= df[df["Country Code"].isin(country_codes)]["Country Code"], 
            z= df[df["Country Code"].isin(country_codes)][year], 
            text= df[df["Country Code"].isin(country_codes)]["Country Name"],
            colorscale = 'Reds',
            autocolorscale = False,
            reversescale = False,
            marker_line_color='darkgray',
            marker_line_width=0.5,
            colorbar_tickprefix = '%',
            colorbar_title = 'Unemployment rate %',
        ))

        fig.update_layout(
            title_text = year + ' unemployment rate',
            geo = dict(
                showframe= False,
                showcoastlines= False,
                projection_type = 'natural earth1'
            )
        )
        
        return fig
```

Visualize highest and lowest unemployment rate in specific years

```python
def generate_BarObject(year):
    year = str(year)
    object = go.Bar(
            x= df[df["Country Code"].isin(country_codes) &  df[year].isin(df[year].nsmallest(5, "first"))].sort_values(year)[year],
            y =df[df["Country Code"].isin(country_codes) &  df[year].isin(df[year].nsmallest(5, "first"))].sort_values(year)["Country Name"],
            orientation= "h",
            )

    return object


def generate_Bars(start_year, period = 1):
    #convert to string
    year2 = start_year + period -1 
    #validate year
    if str(start_year) not in df.columns or str(year2) not in df.columns:
        print("Invalid input")
    else:
        fig = make_subplots(rows = period, 
                            cols = 1, 
                            subplot_titles= list(range(start_year, year2 +1)), 
                            shared_xaxes= True)
        for year in range(start_year,year2+1):
            fig.add_trace(generate_BarObject(str(year)), row= year - start_year +1 , col= 1)

        fig.update_layout(
                title_text = 'Lowest unemployment rate by countries',
                plot_bgcolor="#FFF",  # Sets background color to white
                showlegend = False,    
                xaxis=dict(
                    linecolor="#BCCCDC",  # Sets color of X-axis line
                    showgrid = False,  # Removes X-axis grid lines
                    showspikes=True, # Show spike line for X-axis
                    # Format spike
                    spikethickness=2,
                    spikedash="dot",
                    spikecolor="#999999",
                    spikemode="across",
                ),
                yaxis=dict( 
                    linecolor="#BCCCDC",  # Sets color of Y-axis line
                    showgrid=False,  # Removes Y-axis grid lines    
                ),
            )   
        
        return fig
```
